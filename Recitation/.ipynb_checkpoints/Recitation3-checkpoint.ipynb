```python
import pandas as pd
import numpy as np
import seaborn as sns

df=pd.read_csv('AppleStore.csv')
df.head()
```

```python
types=df["prime_genre"].value_counts().plot(kind="barh")
types.set_title("Apple App Store Applications")
types.set_xlabel("Number of Applications on App Store")
types.set_ylabel("Types of Applications")
```

```python
ax=sns.countplot(data=df, y="prime_genre", order=df["prime_genre"].value_counts(ascending=True).index)
ax.set_title("Apple App Store Applications")
ax.set_xlabel("Number of Applications on App Store")
ax.set_ylabel("Types of Applications")
```

```python
random_array=np.random.randint(6, size=len(df.index), dtype=int)
for i in range(len(df.index)):
    df.loc[i, "Day_of_upload"]=random_array[i]
df
```

```python
df['Day_of_upload']=np.random.randint(0,7,df.shape[0])
```

```python
days=["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
day_indices=range(7)
indices_to_days_dict= dict(zip(day_indices, days))
df["Day_of_upload"]=df["Day_of_upload"].map(indices_to_days_dict)
```

```python
ax=df['Day_of_upload'].value_counts()[days].plot(kind='bar')
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
ax.set_ylabel("Number of Apps For Each Day of the Week")
ax.set_xlabel("Day")
ax.set_ylabel("Number of Apps")
```

```python
count={"0-1":0, "1-2":0, "2-3":0, "3-4":0, "4-5":0}
for i in range(df.shape[0]):
    val=df.loc[i, "user_rating"]
    if val <= 1:
        count["0-1"]+=1
    elif val > 1 and val <=2:
        count["1-2"]+=1
    elif val > 2 and val <=3:
        count["2-3"]+=1
    elif val > 3 and val <=4:
        count["3-4"]+=1
    else:
        count["4-5"]+=1
        
        
new_df=pd.DataFrame()
```
